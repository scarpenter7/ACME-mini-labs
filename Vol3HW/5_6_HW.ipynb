```python
import statsmodels.api as sm
import numpy as np
import pandas as pd
import statsmodels.regression.linear_model as lm
import scipy.linalg as la
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('wages.csv')
df.head(10)
```

```python
df2 = df[['wage', 'female','educ','exper','tenure','married','numdep','nonwhite']]
df2['married*female'] = np.multiply(df2['married'],df2['female'])
df2['constant'] = 1
df2.head(10)
```

```python
def ols(x, y, lam):
    n = len(x.T)
    return la.inv(x.T @ x + n*lam*np.eye(n)) @ x.T @ y
```

```python
ks = range(-5, 11)
print(list(ks))
```

```python
X = df2.drop('wage', axis=1).to_numpy()
y = df2['wage'].to_numpy()

lin = np.linspace(0, 1, 100)

f = lambda betas, x: sum([b*x**i for i,b in enumerate(betas)])
# 5.33a
for k in ks:
    lam = 10**k
    betas = ols(X, y, lam)
    print("lambda = " + str(lam))
    print(betas)
    print()
```

```python
# 5.33b
for k in ks:
    lam = 10**k
    print("lambda = " + str(lam))
    model = lm.OLS(y, X).fit_regularized(alpha=lam, L1_wt=0.0)
    print(model.params)
    print()
```

```python
from sklearn.linear_model import Ridge
# 5.33c
for k in ks:
    lam = 10**k
    print("lambda = " + str(lam))
    model = Ridge(alpha=lam).fit(X, y)
    print(model.coef_)
    print()
```

```python
from sklearn.model_selection import cross_val_score

df = pd.read_csv("Life_Expectancy_Data.csv")
df.head()
```

```python
df['constant'] = 1
```

```python
df.dropna(inplace=True)
```

```python
df = pd.get_dummies(df,columns = ['Country', 'Status'], drop_first=True)
df.head()
```

```python
X = df.drop('Life expectancy ', axis=1).to_numpy()
y = df['Life expectancy '].to_numpy()
#X.head()
```

```python
# y.head()
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error

kf = KFold(n_splits=7)
for k in ks:
    lam = 10**k
    print("lambda = " + str(lam))
    
    mses = []
    
    for train_index, test_index in kf.split(X, y):
        # print("TRAIN:", train_index, "TEST:", test_index)
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
        model = Ridge(alpha = lam).fit(X_train, y_train)
        mse = mean_squared_error(y_test, model.predict(X_test))
        mses.append(mse)
    mean_error = sum(mses)/len(mses)
    print("7 fold CV Score: " + str(mean_error))
    print()
    
    
```

```python
kf = KFold(n_splits=7)

f = lambda betas, x: np.array([np.sum(np.multiply(betas, pt)) for pt in x])

for k in ks:
    lam = 10**k
    print("lambda = " + str(lam))
    
    mses = []
    
    for train_index, test_index in kf.split(X, y):
        # print("TRAIN:", train_index, "TEST:", test_index)
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
        betas = ols(X_train, y_train, lam)
        mse = mean_squared_error(y_test, f(betas, X_test))
        mses.append(mse)
    mean_error = sum(mses)/len(mses)
    print("7 fold CV Score: " + str(mean_error))
    print()
```
