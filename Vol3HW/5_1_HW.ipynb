```python
import numpy as np
import statistics
from matplotlib import pyplot as plt
import pandas as pd
import time
```

```python
# 5.2(i)
def least_squares(X, y):
    sol = np.linalg.solve(X.T@X, X.T@y)
    return sol

n = 100
X = np.hstack([np.ones((n, 1)), np.random.random((n, 1))])
beta = [2, 0.5]
e = .01*np.random.randn(n)
y = np.dot(X, beta) + e
```

```python
res = least_squares(X, y)
print(res)
print(beta)
print(y)
```

```python
# 5.2(ii)
f = lambda b0, b1, x: b0 + b1*x
lin = np.linspace(0, 1, 100)
plt.plot(lin, f(res[0], res[1], lin), label='b_hat')
plt.plot(lin, f(beta[0], beta[1], lin), label='b')
plt.plot(X[:, 1], y, '.')
plt.legend()
plt.show()
```

```python
# 5.2(iv)
n = int(10e6)
X = np.hstack([np.ones((n, 1)), np.random.random((n, 3))])
beta = [1, -2, 3, -4]
e = .5*np.random.randn(n)
y = np.dot(X, beta) + e
res = least_squares(X, y)

f = lambda b0, b1, b2, b3, x: b0 + b1*x + b2*x**2 + b3*x**3
lin = np.linspace(0, 1, 100)
plt.plot(lin, f(res[0], res[1], res[2], res[3], lin), label='b_hat')
plt.plot(lin, f(beta[0], beta[1], beta[2], beta[3], lin), label='b')
#plt.plot(X[:, 1], y, '.')
plt.legend()
plt.show()
print(res)
print(beta)
```

```python
# 5.3
df = pd.read_csv('Datasaurus_Dozen.tsv',sep='\t', header=0)
df.head(10)
```

```python
# set the index to be this and don't drop
df.set_index(keys=['dataset'], drop=False,inplace=True)

# get a list of names
names=df['dataset'].unique().tolist()
print(names)
# now we can perform a lookup on a 'view' of the dataframe
# dino = df.loc[df.dataset=='dino']
for name in names:
    curr_df = df.loc[df.dataset==name]
    x = np.array(curr_df['x'])
    n = len(x)

    stk_x = np.hstack([np.ones((n, 1)), np.array([x]).T])
    x_mean = x.mean()
    x_var = np.var(x)
    print("X Mean: " + str(x_mean))
    print("X Variance: " + str(x_var))

    y = np.array(curr_df['y'])
    y_mean = y.mean()
    y_var = np.var(y)
    print("Y Mean: " + str(y_mean))
    print("Y Variance: " + str(y_var))
    
    res = least_squares(stk_x, y)
    print(res)
    f = lambda b0, b1, x: b0 + b1*x
    lin = np.linspace(0, 100, 100)
    plt.plot(lin, f(res[0], res[1], lin), label='line of best fit')
    plt.plot(x, y, '.')
    plt.legend()
    plt.title(name)
    plt.show()
```

```python
# 5.4 (i)
def svd_inv(x, y):
    U, S, Vh = np.linalg.svd(x)
    n = np.sum(S > 0)
    
    S[:n] = S[:n]**(-1)
    
    x_inv = Vh.T @ np.diag(S) @ U[:, :n].T
    return x_inv @ y
start = time.time()
n = 100
X = np.hstack([np.ones((n, 1)), np.random.random((n, 1))])
beta = [2, 0.5]
e = .01*np.random.randn(n)
y = np.dot(X, beta) + e

res = svd_inv(X, y)
end = time.time()
print(res)
print(beta)
print("SVD time:")
print(end - start)

start = time.time()
n = 100
X = np.hstack([np.ones((n, 1)), np.random.random((n, 1))])
beta = [2, 0.5]
e = .01*np.random.randn(n)
y = np.dot(X, beta) + e

res = least_squares(X, y)
end = time.time()
print(res)
print(beta)
print("Naive time:")
print(end - start)
```
