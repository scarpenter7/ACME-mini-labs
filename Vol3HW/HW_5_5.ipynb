```python
from itertools import chain, combinations
def powerset(iterable):
    "powerset([1,2,3]) --> () (1,) (2,) (3,) (1,2) (1,3) (2,3) (1,2,3)"
    s = list(iterable)
    return chain.from_iterable(combinations(s, r) for r in range(len(s)+1))
```

```python
index_combos = list(powerset(range(8)))[1:]
index_combos = [list(c) for c in index_combos]
print(index_combos)
```

```python
df4 =  df[['wage', 'female','educ','exper','tenure','married','numdep','nonwhite']]
df4['married*female'] = np.multiply(df2['married'],df2['female'])
df4.head()
```

```python
y = df4['wage']
y.head()
```

```python
best_aic = -1
best_combo = None
for c in index_combos:
    X = df4.drop(['wage'], axis=1)
    X = X.iloc[:, c]
    X['constant'] = 1
    results = sm.OLS(y, X).fit()
    aic = results.aic
    if best_aic == -1 or aic < best_aic:
        best_aic = aic
        best_combo = c

print("best combo" + str(best_combo))      
print("best aic: " + str(best_aic))
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
X1 = df4[['female','educ','exper','tenure','married','numdep','nonwhite']]
X1['constant'] = 1
X1['married*female'] = np.multiply(X1['married'],X1['female'])
y = df4['wage']
regr.fit(X1,y)
```

```python
print(regr.coef_)
```

```python
from sklearn.model_selection import train_test_split
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y, test_size=0.3, random_state=42)
```

```python
regr = linear_model.LinearRegression()
regr.fit(X_train1,y_train1)
print(regr.score(X_test1, y_test1))
```

```python
X2 = df4[['educ','tenure','married','married*female']]
X2['constant'] = 1
X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y, test_size=0.3, random_state=42)

regr1 = linear_model.LinearRegression()
regr1.fit(X_train2,y_train2)
print(regr1.score(X_test2, y_test2))
```

```python
from sklearn.model_selection import cross_val_score

scores1 = cross_val_score(regr, X1, y, cv=7)
print(scores1.mean())
scores2 = cross_val_score(regr1, X2, y, cv=7)
print(scores2.mean())
```
